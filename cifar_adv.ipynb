```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10

# Load the CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values to the range [0, 1]
X_test = X_test / 255.0

# Function to generate adversarial perturbations
def generate_adversarial_example(image, epsilon=0.1):
    # Create a random noise matrix with the same shape as the image
    noise = epsilon * np.sign(np.random.randn(*image.shape))
    # Add the noise to the image and clip the values to [0, 1]
    adversarial_image = np.clip(image + noise, 0, 1)
    return adversarial_image

# Choose an image index to visualize
index = 0  # You can change this to visualize different images
clean_image = X_test[index]
perturbed_image = generate_adversarial_example(clean_image, epsilon=0.1)

# Plotting the clean and perturbed images
plt.figure(figsize=(8, 4))

# Display the clean image
plt.subplot(1, 2, 1)
plt.imshow(clean_image)
plt.title("Clean Image")
plt.axis('off')

# Display the perturbed image
plt.subplot(1, 2, 2)
plt.imshow(perturbed_image)
plt.title("Perturbed Image (FGSM)")
plt.axis('off')

plt.show()
```
